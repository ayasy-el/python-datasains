```python
import pandas as pd
import numpy as np
```

```python
# Data Penjualan oleh tiap karyawan
sales = {"Tony": 103, "Sally": 202, "Randy": 380, "Ellen": 101, "Fred": 82}
sales_df = pd.DataFrame.from_dict(sales, orient="index", columns=["sales"])
sales_df
```

```python
# Data Karyawan dengan region nya
region = {
    "Tony": "West",
    "Sally": "South",
    "Carl": "West",
    "Archie": "North",
    "Randy": "East",
    "Ellen": "South",
    "Fred": np.nan,
    "Mo": "East",
    "HanWei": np.nan,
}
region_df = pd.DataFrame.from_dict(region, orient="index", columns=["region"])
region_df
```

```python
join_df = region_df.join(sales_df)
join_df
```

```python
grouped_df = join_df.groupby('region').sum()
grouped_df
```

```python
employee_contrib = join_df.merge(
    grouped_df, how="outer", on="region", suffixes=("", "_total")
).set_index(join_df.index)
```

```python
employee_contrib = employee_contrib.dropna(subset='region')
employee_contrib
```

```python
employee_contrib = employee_contrib.fillna({'sales':0})
employee_contrib
```

```python
employee_contrib['%_of_sales'] = employee_contrib["sales"] / employee_contrib["sales_total"]*100
employee_contrib[["region", "sales", "%_of_sales"]].sort_values(
    by=["region", "%_of_sales"],ascending=False
)
```
