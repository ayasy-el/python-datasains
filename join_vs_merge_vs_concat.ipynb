```python
import pandas as pd

df1 = pd.DataFrame(
    [
        {"A": "A0", "B": "B0", "C": "C0"},
        {"A": "A1", "B": "B1", "C": "C1"},
        {"A": "A2", "B": "B2", "C": "C2"},
        {"A": "A3", "B": "B3", "C": "C3"},
    ],
    index=[0, 1, 2, 3],
)

df1
```

```python
df2 = pd.DataFrame(
    [
        {"B": "B0", "C": "C0", "D": "D0"},
        {"B": "B1", "C": "C1", "D": "D1"},
        {"B": "B2", "C": "C4", "D": "D2"},
        {"B": "B3", "C": "C4", "D": "D3"},
    ],
    index=[2, 3, 4, 5],
)
df2
```

```python
df3 = pd.concat([df1, df2], ignore_index=True, sort=True)
df3
```

```python
df3 = pd.concat([df1, df2], ignore_index=True, sort=False)
df3
```

```python
df1.merge(df2, how="outer")
```

```python
df1.merge(df2, how="cross")
```

```python
left = pd.DataFrame({"k": ["K0", "K1", "K2"], "v": [0, 2, 2]})
right = pd.DataFrame({"k": ["K0", "K1", "K3"], "v": [0, 5, 2]})
left.merge(right, how="outer", on="k")
```

```python
left = pd.DataFrame({"v": [0, 2, 2]}, index=["K0", "K1", "K2"])
right = pd.DataFrame({"x": [0, 5, 2]}, index=["K0", "K1", "K3"])
left.join(right, how="outer")
```

```python
left = pd.DataFrame({"A": [1, 2], "B": [1, 2]})
left
```

```python
right = pd.DataFrame({"A": [4, 5, 6], "B": [2, 2, 2]})
right
```

```python
pd.merge(left, right, on="B", how="outer", validate="1:m")
```
