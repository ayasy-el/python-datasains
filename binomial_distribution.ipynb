```python
import numpy as np
import seaborn as sns
import math
import matplotlib.pyplot as plt

n = 20  # Number of trials (Sample Size)
p = 0.85  # Probability of success on a trial (Success rate)

simulation_count = 100
data = []

for i in range(simulation_count):
    random_numbers = np.random.rand(n)
    x = (random_numbers <= p).sum()
    data.append(x)
    print(random_numbers, x)

sns.set_theme(style="whitegrid", palette="bright")
sns.histplot(data, kde=True, discrete=True)
```

```python
x_values = np.arange(n - 10, n + 2)
binomial_probs = [math.comb(n, k) * (p**k) * ((1 - p) ** (n - k)) for k in x_values]

mu = n * p
sigma = math.sqrt(n * p * (1 - p))

normal_probs = [
    1 / (sigma * math.sqrt(2 * math.pi)) * np.exp(-((x - mu) ** 2) / (2 * sigma**2))
    for x in x_values
]

# Plotting
sns.histplot(data, kde=True, stat="probability", color="royalblue", discrete=True)
sns.lineplot(x=x_values, y=binomial_probs, color="red")
sns.lineplot(x=x_values, y=normal_probs, color="yellow")

plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.title("Binomial Distribution (n={}, p={})".format(n, p))

plt.show()
```

```python
p = 0.85
p2 = 0.70
x_values = np.arange(0, n + 5)

binomial_probs = [math.comb(n, k) * (p**k) * ((1 - p) ** (n - k)) for k in x_values]
binomial_probs2 = [math.comb(n, k) * (p2**k) * ((1 - p2) ** (n - k)) for k in x_values]

sns.lineplot(x=x_values, y=binomial_probs, label="p=0.85")
sns.lineplot(x=x_values, y=binomial_probs2, color="red", label="p=0.70")
```
